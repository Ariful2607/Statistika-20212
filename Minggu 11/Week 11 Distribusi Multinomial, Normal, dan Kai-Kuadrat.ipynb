# Distribusi Multinomial

```python
import numpy as np
from scipy.stats import multinomial
```

Proses pembuatan pensil dalam sebuah pabrik melibatkan banyak buruh dan dilakukan berulang-ulang. Pada suatu pemeriksaan terakhir terlihat bahwa 85% produk adalah ”baik”, 10% ”tidak baik tetapi masih bisa diperbaiki” dan 5% ”rusak dan harus dibuang”. Jika sebuah sampel acak dengan 20 unit dipilih, berapa peluang jumlah unit ”baik” sebanyak 18, unit ”tidak baik tetapi bisa diperbaiki” sebanyak 2 dan unit ”rusak” tidak ada? **Jawaban** = 0,102

```python
# setting nilai n, p, dan x
n = 20
p1 = 0.85
p2 = 0.1
p3 = 0.05
x1 = 18
x2 = 2
x3 = 0
```

```python
multinomial.pmf([x1, x2, x3], n=n, p=[p1, p2, p3])
```

# Distribusi Normal

```python
from scipy.stats import norm
import statistics
```

```python
# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(-1.725, 1.725, 0.01)
```

```python
# Calculating mean and standard deviation
mean = statistics.mean(x_axis)
sd = statistics.stdev(x_axis)
print("Mean = ", mean)
print("Standar Deviasi = ", sd)
```

```python
y_axis = norm.pdf(x_axis, mean, sd)
print (y_axis)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

sb.set_style('whitegrid')
sb.lineplot(x_axis, y_axis , color = 'black')
plt.xlabel('Heights')
plt.ylabel('Probability Density')
```

Luas 𝑃(0 ≤ 𝑍 ≤ 1.20) **jawaban** 0.3849


```python
cdf_upper_limit = norm(loc=mean,scale=sd).cdf(1.2)
cdf_lower_limit = norm(loc=mean,scale=sd).cdf(0)

p = cdf_upper_limit - cdf_lower_limit
print(p)
```

Luas 𝑃(𝑍 ≥ 1,54) **jawaban** 0,0618


```python
cdf_lower_limit = norm(loc=mean,scale=sd).cdf(1.54)

p = 1 - cdf_lower_limit
print(p)
```

Luas 𝑃(𝑍 ≥ -0,86) **jawaban** 0,8051


```python
cdf_lower_limit = norm(loc=mean,scale=sd).cdf(-0.86)

p = 1 - cdf_lower_limit
print(p)
```

Luas 𝑃(0,43 ≤ 𝑍 ≤ 1,12)) **jawaban** 0,2022


```python
cdf_upper_limit = norm(loc=mean,scale=sd).cdf(1.12)
cdf_lower_limit = norm(loc=mean,scale=sd).cdf(0.43)

p = cdf_upper_limit - cdf_lower_limit
print(p)
```

Luas 𝑃(𝑍 ≤0,77) **jawaban** 0,7794


```python
norm(loc=mean,scale=sd).cdf(0.77)
```

# Distribusi Kai-Kuadrat

```python
from scipy.stats import chi2
```

Dengan derajat kebebasan 𝜈=10, luas kurva 𝛼=90%maka luas daerah kurva kai kuadrat berdasarkan tabel terletak di sebelah kanan titik dimana 𝜒 = 16,00 **Jawaban 0,1**


```python
df=10
mean, var, skew, kurt = chi2.stats(df, moments='mvsk')
```

```python
x = np.linspace(chi2.ppf(0.001, df),
                chi2.ppf(0.999, df),90)
y = chi2.pdf(x,df)
print (y)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

sb.set_style('whitegrid')
sb.lineplot(x, y , color = 'black')
plt.xlabel('Heights')
plt.ylabel('Probability Density')
```

```python
cdf_lower_limit = chi2.cdf(16,df)

p = 1 - cdf_lower_limit
print(p)
```
