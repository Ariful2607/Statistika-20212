# Simple Random Sampling

Simple random sampling adalah jenis pengambilan sampel probabilitas dimana memilih secara acak dari sekumpulan set dari suatu populasi.

```python
import random
from random import sample
list1 = [1, 2, 3, 4, 5]
print(sample(list1,3))
```

```python
list1 = [1, 2, 3, 4, 5, 6] 
print("list:", random.sample(list1, 3)) #random sampling list
string = "Statistika"
print("string:", random.sample(string, 5)) #random sampling String
tuple1 = ("UNEJ", "Jember", "FASILKOM", "Sistem Informasi",
                   "Teknologi Informasi", "Informatika", "Mahasiswa")
print("tuple:", random.sample(tuple1, 1)) #random sampling tuple
set1 = {"a", "b", "c", "d", "e"}
print("set:", random.sample(set1, 3)) #random sampling set
```

# Sampling Distribution

```python
import numpy as np
#Definisikan rata-rata dan std dev
mu, sigma = 20, 3
#distribusi normal acak
sample = np.random.normal(mu, sigma, 10)#n=10
print(sample)
```

```python
x_bar = sample.mean()
print(x_bar)
```

```python
import seaborn as sns
#Buat populasi 10000
samples = np.zeros(10000) #N=10000
#Definisikan rata-rata dan std dev
mu, sigma = 20, 3
for s in range(10000):
    #distribusi normal acak
    sample = np.random.normal(mu, sigma, 250)  #n=250
    x_bar = sample.mean()                      #calculating sample mean
    samples[s] = x_bar
g = sns.displot(data=samples, kind="kde")
g.set_axis_labels("Sample Mean", "Density")
```

```python
#Buat populasi 10000
samples = np.zeros(10000) #N=10000
#Definisikan rata-rata dan std dev
mu, sigma = 20, 3
#Definisikan nilai sample
size=[250,100,20,10,5]
for i in size:
    for s in range(10000):
        #distribusi normal acak
        sample = np.random.normal(mu, sigma, i)
        x_bar = sample.mean()                    
        samples[s] = x_bar
    g = sns.displot(data=samples, kind="kde")
    g.set_axis_labels("Sample Mean", "Density")
```

# Prediction Interval

```python
from numpy import mean
from numpy import std
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot as plt
# seed / generate bilangan random
seed(1)
# buat random sampling
x = 20 * randn(1000) + 100
y = x + (10 * randn(1000) + 50)
# mean dan std dari x dan y
print('x: mean=%.3f stdv=%.3f' % (mean(x), std(x)))
print('y: mean=%.3f stdv=%.3f' % (mean(y), std(y)))
# plot
plt.scatter(x, y)
plt.show()
```

Dalam pemodelan statistik, analisis regresi adalah serangkaian proses statistik untuk memperkirakan hubungan antara variabel dependen (sering disebut 'variabel hasil') dan satu atau lebih variabel independen (sering disebut 'prediktor', 'kovariat', atau 'fitur' ). regresi linier adalah pendekatan linier untuk memodelkan hubungan antara respons skalar dan satu atau lebih variabel penjelas (juga dikenal sebagai variabel dependen dan independen).

```python
# linear regression
from scipy.stats import linregress
# seed / generate bilangan random
seed(1)
# buat random sampling
x = 20 * randn(1000) + 100
y = x + (10 * randn(1000) + 50)
# fit model regresi linear
b1, b0, r_value, p_value, std_err = linregress(x, y)
print('b0=%.3f, b1=%.3f' % (b1, b0))
# Prediksi
yhat = b0 + b1 * x
# plot data dan hasil prediksi
plt.scatter(x, y)
plt.plot(x, yhat, color='r')
plt.show()
```
